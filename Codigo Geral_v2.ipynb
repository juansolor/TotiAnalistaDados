# Create Table Warehouse

```python
import sqlite3

# Step 1: Connect to the SQLite database (or create a new one if it doesn't exist)
conn = sqlite3.connect('mydatabase.db')

# Step 2: Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Step 3: Execute SQL commands to create a table
cursor.execute('''
    CREATE TABLE Warehouse (
   Code INTEGER PRIMARY KEY NOT NULL,
   Location TEXT NOT NULL ,
   Quantity INTEGER NOT NULL)
''')


# Step 4: Commit the changes and close the connection
conn.commit()
conn.close()
```

Print Value of the SQL table

```python
# Step 1: Connect to the existing SQLite database
conn = sqlite3.connect('mydatabase.db')

# Step 2: Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Step 3: Execute a SELECT query to fetch data
cursor.execute('SELECT * FROM Warehouse')

# Step 4: Fetch and display the data
rows = cursor.fetchall()

for row in rows:
    print(f"Code: {row[0]}, Location: {row[1]}, Quantity: {row[2]}")

# Step 5: Close the connection
conn.close()
```

SQL lite to DataFrame

```python

# Step 1: Connect to the existing SQLite database
conn = sqlite3.connect('mydatabase.db')

# Step 2: Create a cursor object to execute SQL commands
cursor = conn.cursor()

# Step 3: Execute a SELECT query to fetch data
cursor.execute('SELECT * FROM Warehouse')

# Step 4: Fetch the data and convert it to a DataFrame
data = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]  # Get column names from the cursor description
df = pd.DataFrame(data, columns=columns)

# Step 5: Display the DataFrame
print(df)

# Step 6: Close the connection
conn.close()
```

# Create Table Product

```python
# Step 1: Connect to the SQLite database (or create a new one if it doesn't exist)
conn = sqlite3.connect('mydatabase.db')

# Step 2: Create a cursor object to execute SQL commands
cursor = conn.cursor()


cursor.execute('''
    CREATE TABLE IF NOT EXISTS products (
        Code INTEGER,
        Name VARCHAR(255) NOT NULL,
        Price DECIMAL NOT NULL,
        Warehouse INTEGER NOT NULL,
        PRIMARY KEY (Code),
        FOREIGN KEY (Warehouse) REFERENCES Warehouse (Code)
    )
''')
# Step 4: Commit the changes and close the connection
conn.commit()
conn.close()
```
